```python
#### Tirando nomes

display(df['Descricao'].head(5))

names_df = pd.read_csv('data/nomes-unisex.csv')
lower_names = []
for i in range(0, len(names_df)):
    lower_names.append(names_df.iloc[i]['Nome'].lower())

for i in range(0, len(df)):
    sentence = df.iloc[i]['Descricao'].split()
    for word in sentence:
        if(word.lower() in lower_names):
            df.loc[i, 'Descricao'] = df.iloc[i]['Descricao'].replace(word, '')
```

```python
#Utilizando stemmer e stepwords
stemmer = nltk.stem.RSLPStemmer()
words = nltk.corpus.stopwords.words('portuguese')

for i in range(0, len(df)):
    df.loc[i, 'Descricao'] = normalize('NFKD', df.iloc[i]['Descricao'].decode('utf-8')).encode('ASCII', 'ignore')
        

df['cleaned'] = df['Descricao'].apply(lambda x: " ".join([stemmer.stem(i) for i in re.sub("[^a-zA-Z]", " ", x).split() if i not in words]).lower())

```

```python
import json as j
import pandas as pd
import re
import numpy as np
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2


data = df.copy()



data['cleaned'] = data['Descricao'].apply(lambda x: " ".join([stemmer.stem(i) for i in re.sub("[^a-zA-Z]", " ", x).split() if i not in words]).lower())

X_train, X_test, y_train, y_test = train_test_split(data['cleaned'], data['Categoria'], test_size=0.2)

pipeline = Pipeline([('vect', TfidfVectorizer(ngram_range=(1, 2), stop_words=words, sublinear_tf=True)),
                     ('chi',  SelectKBest(chi2, k=10000)),
                     ('clf', LinearSVC(C=1.0, penalty='l1', max_iter=3000, dual=False))])


model = pipeline.fit(X_train, y_train)

vectorizer = model.named_steps['vect']
chi = model.named_steps['chi']
clf = model.named_steps['clf']

feature_names = vectorizer.get_feature_names()
feature_names = [feature_names[i] for i in chi.get_support(indices=True)]
feature_names = np.asarray(feature_names)

target_names = ['1', '2', '3', '4', '5']
print("top 10 keywords per class:")
for i, label in enumerate(target_names):
    top10 = np.argsort(clf.coef_[i])[-10:]
    print("%s: %s" % (label, " ".join(feature_names[top10])))

print("accuracy score: " + str(model.score(X_test, y_test)))

print(model.predict(['horari host udpaspectap s o window descrica o']))
```
